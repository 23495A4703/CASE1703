```python
import pandas as pd
import matplotlib.pyplot as cr
import numpy as np

balaji=pd.read_csv('energy.csv')
print(balaji)

mumbai=balaji[balaji['State']=='Andhra']
rcb=balaji[balaji["State"]=='Karnataka']

fig,diag=cr.subplots(2)
diag[0].plot(mumbai['Date'],mumbai['Consumption'],color='red')
diag[1].plot(rcb['Date'],rcb['Consumption'])

state1='Kerala'
state2='Delhi'
tdp=[state1,state2]

mumbai=balaji[balaji['State']=='Kerala']['Consumption'].sum()
rcb=balaji[balaji["State"]=='Delhi']['Consumption'].sum()
hari=[mumbai,rcb]
fig,diag=cr.subplots(2)
diag[0].pie(hari,labels=tdp,autopct='%1.1f%%')
```
